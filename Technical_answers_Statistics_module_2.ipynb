
**1. What is a random variable in probability theory?**

A random variable is a variable whose value is a numerical outcome of a random phenomenon.  It's a way to map the outcomes of a random event to numbers.

**2. What are the types of random variables?**

There are two main types:

* **Discrete Random Variables:**  Can only take on a finite number of values or a countably infinite number of values (e.g., the number of heads in 10 coin flips, the number of cars passing a point on a road in an hour).
* **Continuous Random Variables:** Can take on any value within a given range (e.g., a person's height, the temperature of a room).

**3. What is the difference between discrete and continuous distributions?**

* **Discrete Distributions:** Describe the probabilities of discrete random variables.  Probabilities are assigned to specific values.  The sum of all probabilities must equal 1.
* **Continuous Distributions:** Describe the probabilities of continuous random variables. Probabilities are assigned to ranges of values.  The probability at a single point is technically zero.  The area under the curve (the probability density function) represents probability. The total area under the curve is 1.

**4. What are probability distribution functions (PDFs)?**

* **Probability Mass Function (PMF):** For discrete random variables. Gives the probability of the variable taking on a specific value.
* **Probability Density Function (PDF):** For continuous random variables.  The integral of the PDF over an interval gives the probability of the variable falling within that interval. The PDF itself doesn't directly give probabilities.

**5. How do cumulative distribution functions (CDFs) differ from probability distribution functions (PDFs)?**

* **PDF/PMF:** Gives the probability of a specific value or the probability density at a point.
* **CDF:** Gives the probability that a random variable is *less than or equal to* a given value.  It's the cumulative probability up to that point.

**6. What is a discrete uniform distribution?**

A discrete uniform distribution is a distribution where each of a finite number of values has an equal probability of being selected.  Like rolling a fair die – each face has a 1/6 probability.

**7. What are the key properties of a Bernoulli distribution?**

* Represents the probability of success or failure of a single trial.
* Has two possible outcomes (usually 0 and 1).
* Characterized by a single parameter, *p*, the probability of success.

**8. What is the binomial distribution, and how is it used in probability?**

The binomial distribution describes the probability of getting exactly *k* successes in *n* independent Bernoulli trials (each with the same probability of success, *p*). It's used when you have repeated trials with two outcomes.

**9. What is the Poisson distribution and where is it applied?**

The Poisson distribution describes the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known average rate and independently of the time since the last event. It's used for modeling rare events (e.g., number of phone calls at a call center per minute, number of cars passing a point on a road per hour).

**10. What is a continuous uniform distribution?**

A continuous uniform distribution is a distribution where all values within a given interval are equally likely.

**11. What are the characteristics of a normal distribution?**

* Bell-shaped and symmetrical.
* Defined by two parameters: the mean (μ) and the standard deviation (σ).
* Many natural phenomena follow a normal distribution or can be approximated by one.

**12. What is the standard normal distribution, and why is it important?**

The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It's important because:

* Any normal distribution can be transformed into a standard normal distribution (by standardizing).
* Probabilities for the standard normal distribution are readily available in tables or can be calculated using software.

**13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

The CLT states that the distribution of the *sample means* of a large number of independent and identically distributed random variables will approach a normal distribution, regardless of the shape of the original distribution.  It's critical because it allows us to use normal distribution theory for hypothesis testing and confidence intervals, even if the underlying population is not normally distributed, as long as the sample size is large enough.

**14. How does the Central Limit Theorem relate to the normal distribution?**

The CLT says that sample means will be *approximately* normally distributed as the sample size increases.

**15. What is the application of Z statistics in hypothesis testing?**

Z-statistics are used to test hypotheses about population means when the population standard deviation is known or the sample size is large. They compare the sample mean to a hypothesized population mean in terms of standard errors.

**16. How do you calculate a Z-score, and what does it represent?**

Z-score = (data point - population mean) / population standard deviation

It represents how many standard deviations a data point is away from the mean.

**17. What are point estimates and interval estimates in statistics?**

* **Point Estimate:** A single value estimate of a population parameter (e.g., the sample mean as an estimate of the population mean).
* **Interval Estimate:** A range of values within which a population parameter is likely to fall (e.g., a confidence interval).

**18. What is the significance of confidence intervals in statistical analysis?**

Confidence intervals provide a range of plausible values for a population parameter, along with a level of confidence that the true parameter lies within that range. They quantify the uncertainty associated with estimating population parameters from samples.

**19. What is the relationship between a Z-score and a confidence interval?**

Z-scores are used to calculate confidence intervals when the population standard deviation is known.  The Z-score corresponding to the desired confidence level determines the margin of error for the interval.

**20. How are Z-scores used to compare different distributions?**

Z-scores standardize data, making it easier to compare values from different distributions. By converting data points to Z-scores, you can see how far each point is from its respective mean in terms of standard deviations.

**21. What are the assumptions for applying the Central Limit Theorem?**

* The samples must be random and independent.
* The sample size should be sufficiently large (generally considered to be at least 30).  The larger the sample size, the better the normal approximation.
* The original population should have a finite variance.

**22. What is the concept of expected value in a probability distribution?**

The expected value (or mean) of a random variable is the average value you would expect if you repeated the random experiment many times. It's a weighted average of all possible values, where the weights are the probabilities.

**23. How does a probability distribution relate to the expected outcome of a random variable?**

The probability distribution describes the likelihood of each possible outcome of a random variable. The expected value is calculated using the probabilities from the distribution. The expected value is the long-run average outcome we expect to see.
